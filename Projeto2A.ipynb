## Imagens digitais

Uma *imagem digital* pode ser vista como uma matriz $\mathbf{P} = \left[ p_{ij} \right]$, $m\times n$, onde cada elemento $p_{ij}$  denominado *pixel*.
Os cantos superior esquerdo e inferior direito de uma imagem correspondem aos pixels $p_{0,0}$ e $p_{m-1,n-1}$.
Existem diversas formas de se representar a cor de um pixel.
No sistema conhecido como RGB (do ingls, *Red*, *Green*, *Blue*), a cor de um pixel  representada usando trs canais: <font color="red">vermelho</font>, <font color="green">verde</font> e <font color="blue">azul</font>.
Supondo cada canal representado com 8 bits, teremos $2^8 = 256$ valores distintos por canal.
A combinao dos trs canais possibilita a criao de 256 x 256 x 256 cores, o que  algo em torno de 16 milhes de cores.

Matematicamente, a cor de um pixel $p_{i,j}$ pode ser representada por um vetor de cor $p_{i,j} = \left( r, g, b\right)$, onde $0 \le r,g,b \le 255$.
Quanto maior o valor de um canal, maior ser a intensidade dele na cor final obtida.
Algumas exemplos de combinaes de canais so:

* $(255,0,0)$: <font color="#ff0000">vermelho</font>
* $(0,255,0)$: <font color="#00ff00">verde</font>
* $(0,0,255)$: <font color="#0000ff">azul</font>
* $(255,255,0)$: <font color="#ffff00">amarelo</font>
* $(0,255,255)$: <font color="#00ffff">ciano</font>
* $(255,0,255)$ <font color="#ff00ff">magenta</font>
* $(0,0,0)$: <font color="#000000">preto</font>
* $(127,127,127)$: <font color="#777777">cinza mdio</font>
* $(255,255,255)$: <font color="#ffffff">branco</font> (no d para ver, no ? Selecione o texto ao lado.)

## Algoritmo de retoque

O objetivo de um *algoritmo de retoque*  substituir pixels indesejados de uma foto baseando-se nas cores de seus pixels vizinhos.
Um bom retoque deve produzir uma imagem nova com emendas imperceptveis entre as regies que foram mantidas intactas e aquelas recm preenchidas.
A imagem abaixo ilustra o resultado de um algoritmo de retoque.
 esquerda est a imagem original, ao centro esto indicados os pixels que se deseja alterar, e ao lado direito est o resultado final.

![](https://raw.githubusercontent.com/Mugichoko445/Fast-Digital-Image-Inpainting/images/example_2.png)
Fonte: https://github.com/Mugichoko445/Fast-Digital-Image-Inpainting

Um algoritmo simples de retoque funciona do seguinte modo.
Suponha que $\mathbf{P} = \left[p_{i,j}\right]$  a imagem original.
Primeiramente, vocs devem indicar quais pixels de $\mathbf{P}$ desejam retocar.
Esta fase  a que toma mais tempo, pois ainda  feita manualmente.
Para tanto, criamos uma imagem auxiliar $\mathbf{M} = [m_{i,j}]$, conhecida como *mscara*, com as mesmas dimenses de $\mathbf{P}$.
Nela, indicamos em preto os pixels que desemamos remover.
Se a cor armazenada em $m_{i,j}$ for diferente de preto, ento o pixel $p_{i,j}$ dever ser preservado e nada precisa ser feito com relao a ele.
Caso contrrio, sua cor ser recalculada como a seguir.

Suponha que $p_{i,j}$  um pixel que precisa ser alterado, ou seja, $m_{i,j} = (0,0,0)$.
Portanto, sua vizinhana  forma pelos pixels:

$$
\begin{array}{|c|c|c|}
\hline
p_{i-1,j-1} & p_{i-1,j} & p_{i-1,j+1}\\
\hline
p_{i,j-1} & p_{i,j} & p_{i,j+1}\\
\hline
p_{i+1,j-1} & p_{i+1,j} & p_{i+1,j+1}\\
\hline
\end{array}
$$

A ideia bsica do algoritmo  expressar a nova cor de $p_{i,j}$ como a mdia aritmtica das cores de seus vizinhos, isto ,

\begin{multline*}
(r_{i,j},g_{i,j},b_{i,j}) = \frac{1}{8} \left[ (r_{i-1,j-1},g_{i-1,j-1},b_{i-1,j-1}) + (r_{i-1,j},g_{i-1,j},b_{i-1,j}) + (r_{i-1,j+1},g_{i-1,j+1},b_{i-1,j+1}) + (r_{i,j-1},g_{i,j-1},b_{i,j-1}) + (r_{i,j+1},g_{i,j+1},b_{i,j+1}) + (r_{i+1,j-1},g_{i+1,j-1},b_{i+1,j-1}) + (r_{i+1,j},g_{i+1,j},b_{i+1,j}) + (r_{i+1,j+1},g_{i+1,j+1},b_{i+1,j+1}) \right]
\end{multline*}

Por convenincia, considere apenas o canal R (vermelho). Nesse caso, podemos escrever:

$$
-r_{i-1,j-1} - r_{i-1,j} - r_{i-1,j+1} - r_{i,j-1} + 8 r_{i,j} - r_{i,j+1} - r_{i+1,j-1} - r_{i+1,j} - r_{i+1,j+1} = 0
$$

Se um dos pixels vizinhos de $p_{i,j}$ precisar ser preservado, sua cor deve ser tratada como constante. Neste caso, esta parcela dever ser transferida para a direita da equao.

Supondo um total de $N$ pixels indesejados, a aplicao da equao acima em cada um deles resultar em trs sistemas lineares com $N$ equaes e $N$ incgnitas, um para cada canal RGB, todos com a mesma matriz de coeficientes:

$$
\mathbf{A}\mathbf{x}_r = \mathbf{b}_r, \quad
\mathbf{A}\mathbf{x}_g = \mathbf{b}_g, \quad \text{e} \quad
\mathbf{A}\mathbf{x}_b = \mathbf{b}_b.
$$

Aps resolver esses sistemas, basta atualizar as cores dos pixels indesejados com aquelas que foram calculadas e voil, sua imagem ter sido retocada!

Com efeito, para o primeiro pixel indesejado $p_{1,2}$, temos as equao

$$
-r_{0,1} - r_{0,2} - r_{0,3} - r_{1,0} + 8 r_{1,2} - r_{1,3} - r_{2,0} - r_{2,2} - r_{2,3} = 0
$$

J para o pixel $p_{2,2}$, obtemos

$$
-r_{1,1} - r_{1,2} - r_{1,3} - r_{2,1} + 8 r_{2,2} - r_{2,3} - r_{3,1} - r_{3,2} - r_{3,3} = 0

Como todos os demais pixels devem ser preservados (so constantes), podemos reescrever as equaes acima como:

\begin{align*}
8 r_{1,2} - r_{2,2} &= r_{0,1} + r_{0,2} + r_{0,3} + r_{1,0} + r_{1,3} + r_{2,0} + r_{2,3}\\

\end{align*}

que matricialmente fica:

$$
\begin{bmatrix}
 8 & - 1\\
-1 & 8
\end{bmatrix}
\begin{bmatrix}
r_{1,2}\\
r_{2,2}
\end{bmatrix}
=
\begin{bmatrix}
r_{0,1} + r_{0,2} + r_{0,3} + r_{1,0} + r_{1,3} + r_{2,0} + r_{2,3}\\
r_{1,1} + r_{1,3} + r_{2,1} + r_{2,3} + r_{3,1} + r_{3,2} + r_{3,3}
\end{bmatrix}
$$

Procedendo de modo anlogo, montamos os sistemas para os canais G e B:

$$
\begin{bmatrix}

-1 & 8
\end{bmatrix}
\begin{bmatrix}

\end{bmatrix}
=
\begin{bmatrix}
g_{0,1} + g_{0,2} + g_{0,3} + g_{1,0} + g_{1,3} + g_{2,0} + g_{2,3}\\
g_{1,1} + g_{1,3} + g_{2,1} + g_{2,3} + g_{3,1} + g_{3,2} + g_{3,3}
\end{bmatrix}
$$

$$
\begin{bmatrix}
 8 & - 1\\
-1 & 8
\end{bmatrix}
\begin{bmatrix}
b_{1,2}\\
b_{2,2}
\end{bmatrix}
=
\begin{bmatrix}
b_{0,1} + b_{0,2} + b_{0,3} + b_{1,0} + b_{1,3} + b_{2,0} + b_{2,3}\\
b_{1,1} + b_{1,3} + b_{2,1} + b_{2,3} + b_{3,1} + b_{3,2} + b_{3,3}
\end{bmatrix}
$$

Resolvendo esses sistemas e atualizando a imagem com os novos valores de cores calculados para os pixels indesejados, teremos a imagem retocada.
